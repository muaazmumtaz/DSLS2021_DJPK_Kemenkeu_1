PUNCTUATION REMOVAL

```python
import string
import re

def remove_punctuation(text):
    return text.translate(str.maketrans("","",string.punctuation))

# Pada Nama Kegiatan
df['Namakegiatan'] = df['Namakegiatan'].apply(remove_punctuation)
df1['Namakegiatan'] = df1['Namakegiatan'].apply(remove_punctuation)
df2['Namakegiatan'] = df2['Namakegiatan'].apply(remove_punctuation)
df3['Namakegiatan'] = df3['Namakegiatan'].apply(remove_punctuation)
df4['Namakegiatan'] = df4['Namakegiatan'].apply(remove_punctuation)
```

```python
# Pada Nama Kegiatan
df['Namakegiatan'] = df['Namakegiatan'].apply(lambda x: [i for i in x if i not in stop_words])
df1['Namakegiatan'] = df1['Namakegiatan'].apply(lambda x: [i for i in x if i not in stop_words])
df2['Namakegiatan'] = df2['Namakegiatan'].apply(lambda x: [i for i in x if i not in stop_words])
df3['Namakegiatan'] = df3['Namakegiatan'].apply(lambda x: [i for i in x if i not in stop_words])
df4['Namakegiatan'] = df4['Namakegiatan'].apply(lambda x: [i for i in x if i not in stop_words])
```

```python
# join list of token as single document string
import ast

def join_text_list(texts):
    texts = ast.literal_eval(texts)
    return ' '.join([text for text in texts])

df["Namakegiatan_stem"] = df["Namakegiatan_stem"].apply(join_text_list)
df["Namakegiatan"] = df["Namakegiatan"].apply(join_text_list)
df1["Namakegiatan_stem"] = df1["Namakegiatan_stem"].apply(join_text_list)
df1["Namakegiatan"] = df1["Namakegiatan"].apply(join_text_list)
df2["Namakegiatan_stem"] = df2["Namakegiatan_stem"].apply(join_text_list)
df2["Namakegiatan"] = df2["Namakegiatan"].apply(join_text_list)
df3["Namakegiatan_stem"] = df3["Namakegiatan_stem"].apply(join_text_list)
df3["Namakegiatan"] = df3["Namakegiatan"].apply(join_text_list)
df4["Namakegiatan_stem"] = df4["Namakegiatan_stem"].apply(join_text_list)
df4["Namakegiatan"] = df4["Namakegiatan"].apply(join_text_list)
```
